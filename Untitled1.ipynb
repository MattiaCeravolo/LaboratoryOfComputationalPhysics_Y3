```python
import matplotlib.pyplot as plt
%matplotlib inline

# always useful
import numpy as np
from scipy import stats
from math import pi
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_title('Curva di Lorenz',fontsize=16)
ax.set_ylabel('Frazione di ricchezza Y(X)',fontsize=16)
ax.set_xlabel('Frazione di popolazione X',fontsize=16)
data = np.sort(np.random.gamma(1,size=10000))
x = np.arange(1,data.size+1)/data.size
y = np.cumsum(data)/data.sum()
ax.plot(x,y,c = 'black', lw = 3, label = "Distribuzione esponenziale $e^{-x}$")
ax.plot(np.linspace(0,1,10),np.linspace(0,1,10),linestyle = 'dashed', c  = 'black', lw = 3, label = "Curva di uguaglianza")
ax.legend(loc = 'best')
x0 = x[(x-y).argmax()]
y0 = y[(x-y).argmax()]
ax.plot([x0,x0],[y0,x0], c = 'red', lw = 3)
ax.annotate('$\max_{X \in [0,1]} (L(X) - L_{pe}(X)) = 0.37$', xy=(x0, (y0 + x0)/2), xytext=(x0-0.1, 0.15),
            arrowprops=dict(facecolor='black', shrink=0.05),fontsize=16
            )
```
