```python
import matplotlib.pyplot as plt
%matplotlib inline

# always useful
import numpy as np
from scipy import stats
from math import pi
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_ylabel('$Y(m)$',fontsize=24)
ax.set_xlabel('$X(m)$',fontsize=24)
data = np.sort(np.random.gamma(1,size=10000))
x = np.arange(1,data.size+1)/data.size
y = np.cumsum(data)/data.sum()
ax.plot(x,y,c = 'black', lw = 3, label = "Curva di Lorenz $L(X) = Y(m^{-1}(X))$")
ax.plot(np.linspace(0,1,10),np.linspace(0,1,10),linestyle = 'dashed', c  = 'black', lw = 3, label = "Curva di uguaglianza $L_{pe}(X)=X$")
ax.legend(loc = 'best', fontsize = 24)
ax.grid()
ax.tick_params(axis="x", labelsize=24)
ax.tick_params(axis="y", labelsize=24)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_ylabel('$Y(m)$',fontsize=24)
ax.set_xlabel('$X(m)$',fontsize=24)
data = np.sort(np.random.gamma(1,size=10000))
x = np.arange(1,data.size+1)/data.size
y = np.cumsum(data)/data.sum()
ax.plot(x,y,c = 'black', lw = 3, label = "Curva di Lorenz $L(X) = Y(m^{-1}(X))$")
ax.plot(np.linspace(0,1,10),np.linspace(0,1,10),linestyle = 'dashed', c  = 'black', lw = 3, label = "Curva di uguaglianza $L_{pe}(X)=X$")
ax.plot(x,1-x, c  = 'blue', lw = 3, label = "$Y=1-X$")
ax.legend(loc = 'best', fontsize = 24)
ax.grid()
ax.tick_params(axis="x", labelsize=24)
ax.tick_params(axis="y", labelsize=24)

x0 = x[(x-y).argmax()]
y0 = y[(x-y).argmax()]
ax.plot([x0,x0],[y0,x0], c = 'red', lw = 3)
ax.annotate('$\max_{X \in [0,1]} (L(X) - L_{pe}(X))$', xy=(x0, (y0 + x0)/2), xytext=(x0-0.6, 0.65),
            arrowprops=dict(facecolor='black', shrink=0.05),fontsize=24
            )

x1 = x[np.absolute(y -(1-x)).argmin()]
y1 = y[np.absolute(y -(1-x)).argmin()]
ax.plot([x1,x1],[0,y1], c = 'blue', lw = 3)
ax.annotate('$k = X*$', xy=(x1, 0), xytext=(x0+0.2, 0.35),
            arrowprops=dict(facecolor='black', shrink=0.05),fontsize=24
            )
```
